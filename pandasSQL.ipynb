Here I used pandas, sqldf, random libraries for creating dataframe 

```python
import pandas as pd
from pandasql import sqldf
from random import choice, random
```

Dataframe has four columns with random choice.

```python
major_choices=['math','Physics','computer science','english','history','biology','chemistry']
year_choices=[1,2,3,4,4]
major_gpa_avgs={major:random()+2.5 for major in major_choices}
n=100000
names=[''.join([choice('abcdefghijklmnopqrstuvwxyz') for _ in range(7)]) for _ in range(n)]
major=[choice(major_choices) for _ in range(n)]
gpas=[min(major_gpa_avgs[m]+random()/4.0,4.0)for m in major]
years=[choice(year_choices) for _ in range(n)]

student_data=pd.DataFrame(data=zip(names,gpas,major,years),columns=['Names','GPA','Major','Year'])
```

# SQLite3 to create a table and insert values into table.

```python
import sqlite3

def display_like_table(data,headers):
    df=pd.DataFrame(data=data,columns=[i[0] for i in headers])
    return df
```

```python
#create a database connection
conn=sqlite3.connect('student.db')
```

```python
cur=conn.cursor()
```

```python
#create a student table
cur.execute("create table students(Name text,GPA float,Major text, Year integer)")
```

```python
#display the student table
cur.execute("select * from students")
display_like_table(cur.fetchall(), cur.description)
```

```python
#Insert a row into table
cur.execute("insert into students values('praveen',2.3,'math',2)")
cur.execute("select * from students")
display_like_table(cur.fetchall(),cur.description)
```

```python
conn.commit()
```
